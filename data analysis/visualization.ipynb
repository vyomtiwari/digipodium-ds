```python
x = np.arange(-5, 5, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
figs, axs = plt.subplots(nrows =1, ncols=3, figsize =(10,3))
axs[0].plot(x, y1, 'r--')
axs[1].plot(x, y2, 'b--')
axs[2].plot(x, y3, 'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')

```

implicit vs explicit

```python
# plt -> implicit figure and 
x = np.linspace(-10, 10, 25)
plt.plot(x, x, '-', label = 'linear')
plt.plot(x, x**2, '--', label = 'quadratic')
plt.plot(x, x**3, 'o--', label ='cubic')
plt.xlabel('my x axis')
plt.ylabel('my y label')
plt.title('simple lot with implicit graph setting')
plt.legend()
plt.show()
```

```python
# explicit
x = np.linspace(-10, 10, 25)
fig, ax = plt.subplots() # fig and ax are objects
ax.plot(x, x, '-', label = 'linear')
ax.plot(x, x**2, '--', label = 'quadratic')
ax.plot(x, x**3, 'o--', label ='cubic')
ax.set_xlabel('my x axis')
ax.set_ylabel('my y label')
ax.set_title('simple lot with implicit graph setting')
ax.legend()
plt.show()
```

```python
# usign dictionary for graph 
data = {
    'a':np.arange(50),
    'b': np.random.randint(0, 50, 50),
    'c': np.random.randn(50),
    'd': np.arange(50)+10*np.random.randn(50)
}
data['d']= np.abs(data['d']*10)
fig, ax = plt.subplots(figsize = (10,3))
ax.scatter('a', 'b', c = 'c', s = 'd', data = data)
ax.set_xlabel('entry a')
ax.set_ylabel('enter b')
ax.set_title('Scatter Plot')
plt.show()
```

```python
data = {
    'names': ['Robert', 'essica', 'Margraet', 'Johnthan', 'Melbourne'],
    'english':[76, 95, 77, 78, 39],
    'math':[73, 86, 57, 18, 29],
    'hindi':[71, 81, 57, 78, 39]
      }
```

```python
def my_custom_plotter(ax, x_value, y_value, param_dict):
    out = ax.bar(x_value, y_value, **param_dict)
    return out
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,3))
my_custom_plotter(axs[0], 'names', 'english', {'data':data, 'color':'red', 'alpha':.8})
my_custom_plotter(axs[1], 'names', 'hindi', {'data':data, 'color':'black', 'alpha':.9})
my_custom_plotter(axs[2], 'names', 'math', {'data':data,'color':'blue', 'alpha':.2})
# saving the graph as image
plt.savefig('Plot1.png', bbox_inches = 'tight', dpi = 300)
```
